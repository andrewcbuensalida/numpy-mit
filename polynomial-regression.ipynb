```python
# linear regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Training data
X = np.random.rand(100, 1)
# print(X)
y = 4 + 5 * X + np.random.randn(100, 1) # add noise

reg = LinearRegression()
reg.fit(X, y) # this creates the model

X_vals = np.linspace(0, 1, 100).reshape(-1, 1) # -1 infers the correct dimensions based on the other dimensions. This is testing data.
y_vals = reg.predict(X_vals)

plt.scatter(X,y)
plt.plot(X_vals, y_vals, color='red') # since didn't use show yet, it will combine this line plot with the scatter plot above
plt.show() # after show, it clears the plot
```

```python
# now for polynomial regression
from sklearn.preprocessing import PolynomialFeatures

# Training data
X = 4 * np.random.rand(100, 1) - 2
# print(X)
y = 4 + 2 * X + 5 * X**2 + np.random.randn(100, 1)  # add noise

poly_features = PolynomialFeatures(degree=2, include_bias=False) # degree 2 means quadratic. trick here is to not make this too high or else it will overfit, especially with lots of noise, meaning it will fit the training data too well and not generalize well to new data
X_poly = poly_features.fit_transform(X)

reg = LinearRegression()
reg.fit(X_poly, y)  # this creates the model

X_vals = np.linspace(-2, 2, 100).reshape(
    -1, 1
)  # -1 infers the correct dimensions based on the other dimensions
X_vals_poly = poly_features.transform(X_vals)
y_vals = reg.predict(X_vals_poly)

plt.scatter(X, y)
plt.plot(
    X_vals, y_vals, color="red"
)  # since didn't use show yet, it will combine this line plot with the scatter plot above
plt.show()  # after show, it clears the plot
```
